## Galton Board

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from collections import Counter 

def makeGalton(kugeln=10000, marbels=10, prob=[0.5,0.5]):
    
    nums = []
    # für jede Kugel wird Verlauf durchgespielt
    for i in range(kugeln):
        num = 0
        for j in range(1,marbels + 1):

            lr = np.random.choice(np.arange(0, 2), p=prob)
            num += lr
        nums.append(num)
    
    nums_counts = Counter(nums)  
    counter = sorted(nums_counts.items())
    # Print Daten
    for x, y in counter:
        print("Slot: {} Anzahl an Kugeln: {}".format(x, y))
    
    # plot Histogram und Dichtefunktion
    bins = np.arange(marbels + 1) - 0.5
    
    ax = sns.distplot(nums, hist = True,fit=norm, bins=bins, kde=False, hist_kws=dict(edgecolor="k", linewidth=2))
    plt.title('Galton Board für {} Kugeln'.format(kugeln))
    plt.xlabel('Slot')
    plt.xticks(range(marbels + 1))
    plt.ylabel('Dichte')
    plt.show()
```

```python
## Fall wenn die Wahrscheinlichkeit ob Kugel links oder rechts fällt ist jeweils 50% ist
makeGalton(kugeln=10000, marbels=10, prob=[0.5,0.5])
```

```python
## Fall wenn die Wahrscheinlichkeit ob Kugel links oder rechts fällt 90% links bzw. 10% rechts ist
makeGalton(kugeln=10000, marbels=10, prob=[0.9,0.1])
```

```python
## Fall wenn die Wahrscheinlichkeit ob Kugel links oder rechts fällt 70% links bzw. 30% rechts ist
makeGalton(kugeln=10000, marbels=10, prob=[0.7,0.3])
```

```python
## Fall mit 100 Kugeln
makeGalton(kugeln=100, marbels=10, prob=[0.5,0.5])
```
